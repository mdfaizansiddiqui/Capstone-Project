```python
def change_to_time_index(df):
    df['Date2'] = pd.to_datetime(df['Date'])
    return df
```

```python
amazon_senti = change_to_time_index(amazon_senti)
amazon_stock_trend = change_to_time_index(amazon_stock_trend)
```

```python
amazon_senti_daily = amazon_senti.groupby(pd.Grouper(key='Date2',freq='1D')).mean()
```

```python
def combine(stock_df, sentiment_df):
    target_variables = ['Adj Close', 'Target_Direction', 'Date']
    stock_trend = stock_df[target_variables]
    senti = sentiment_df[['Date', 'sentiment']]
    senti = change_to_time_index(senti)
    stock_trend = change_to_time_index(stock_trend)
    
    stock_trend.drop(['Date2'], axis = 1, inplace = True)
    stock_trend['Date'] = pd.to_datetime(stock_trend['Date'])
    
    senti_daily = senti.groupby(pd.Grouper(key='Date2',freq='1D')).mean()
    senti_daily['Date'] = senti_daily.index
    merge1 = pd.merge(senti_daily, stock_trend,  on='Date', how ='left')
    dates = pd.to_datetime(stock_df['Date'])
    merge2 = pd.merge(merge1, dates, on='Date')
    merge2['sentiment'] = merge2['sentiment'].apply(lambda x : round(x, 1))
    merge2['Adj Close'] = merge2['Adj Close'].apply(lambda x : round(x, 4))
    return merge2
```

```python
amazon_merge = combine(amazon_2k13, amz)
```

```python
apple_merge = combine(apple_2k13, appl)
```
